```python
import cv2
```

```python
from keras.models import load_model 
import keras
import pyautogui
import numpy as np
import pandas as pd
import urllib3
import threading
import time
```

```python
model = load_model(r"d:\Users\Shrikaran\Desktop\Project AI-CV\Scripts\model\resnetmodel.hdf5")
```

```python
# url = 'https://192.168.1.7:8080/video'
# cap = cv2.VideoCapture(url)

vid = cv2.VideoCapture('http://192.168.1.100:8080/video')
vid.set(cv2.CAP_PROP_FRAME_WIDTH,400)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT,400)
```

```python
labels = pd.read_csv(r"d:\Users\Shrikaran\Desktop\Project AI-CV\dataset\jester-v1-labels.csv",header = None)
```

```python
buffer = []
cls = []
predicted_value = 0
final_label = ""
i = 1
while(vid.isOpened()):
    ret, frame = vid.read()
#     print(ret)
#     print(frame)
    if ret:
        image = cv2.resize(frame, (96,64))
        image = image/255.0
        buffer.append(image)
        if(i%16 == 0):
            buffer = np.expand_dims(buffer,0)
            predicted_value = np.argmax(model.predict(buffer))
            cls = labels.iloc[predicted_value]
#             print(predicted_value)
#             print(cls)
#             print(cls.iloc[0])
            if(predicted_value == 0):
                final_label = "Swipping Left"
                pyautogui.press('p') #goto previous 
                time.sleep(.15)
                
            elif(predicted_value ==1):
                final_label="Swiping Right"
                pyautogui.press('n') #goto next
                time.sleep(.15)
                
            elif(predicted_value ==2):
                final_label ="Swiping Down"
                pyautogui.press('volumedown') #to decrease volume by 2
                time.sleep(.15)
                
            elif(predicted_value ==3):
                final_label ="Swiping Up"
                pyautogui.press('volumeup') #t0 increase volume by 2
                time.sleep(.15)
                
            elif(predicted_value ==4):
                final_label ="Pushing Hand Away"
                pyautogui.press(']') #to make faster rate
                time.sleep(.15)
                
            elif(predicted_value ==5):
                final_label ="Pulling Hand In"
                pyautogui.press('[') #to make slower rate
                time.sleep(.15)
                
            elif(predicted_value ==6):
                final_label ="Sliding Two Fingers Left"
                pyautogui.press('Left') #skip 10 seconds backward
                time.sleep(.15)
                
            elif(predicted_value ==7):
                final_label ="Sliding Two Fingers Right"
                pyautogui.press('Right') #skip 10 seconds forward
                time.sleep(.15)
                
            elif(predicted_value ==8):
                final_label ="Sliding Two Fingers Down"
                pyautogui.press('Down') #to decrease volume by 5
                time.sleep(.15)
                
            elif(predicted_value ==9):
                final_label ="Sliding Two Fingers Up"
                pyautogui.press('Up') #to increase volume by 5
                time.sleep(.15)
                
            elif(predicted_value ==10):
                final_label ="Pushing Two Fingers Away"
                pyautogui.press('Space') #to play
                time.sleep(.15)
                
            elif(predicted_value ==11):
                final_label ="Pulling Two Fingers In"
                pyautogui.hotkey('ctrl', 'q') # to quit application
                time.sleep(.15)
                
            elif(predicted_value ==12):
                final_label ="Rolling Hand Forward"
                pyautogui.hotkey('ctrl', 'Right') #to skip 1_minute forward
                time.sleep(.15)
                
            elif(predicted_value ==13):
                final_label ="Rolling Hand Backward"
                pyautogui.hotkey('ctrl', 'Left') #to skip 1_minute backward
                time.sleep(.15)
                
            elif(predicted_value ==14):
                final_label ="Turning Hand Clockwise"
                pyautogui.hotkey('alt', 'ctrl', 'Right') #to skip 5_minutes forward
                time.sleep(.15)
                
            elif(predicted_value ==15):
                final_label ="Turning Hand Counterclockwise"
                pyautogui.hotkey('alt', 'ctrl', 'Left') #to skip 5_minutes backward
                time.sleep(.15)
                                
            elif(predicted_value ==16):
                final_label ="Zooming In With Full Hand"         
                pyautogui.hotkey('shift', 'Right') #to skip 3_seconds forward
                time.sleep(.15)
                
            elif(predicted_value ==17):
                final_label ="Zooming Out With Full Hand"
                pyautogui.hotkey('shift', 'Left') #to skip 3_seconds backward
                time.sleep(.15)
                
            elif(predicted_value ==18):
                final_label ="Zooming In With Two Fingers"
                pyautogui.hotkey('ctrl', '+') #zoom in picture  
                time.sleep(.15)
                
            elif(predicted_value ==19):
                final_label ="Zooming Out With Two Fingers"
                pyautogui.hotkey('ctrl', '-') #zoom out picture
                time.sleep(.15)
                
            elif(predicted_value ==20):
                final_label ="Thumb Up"
                pyautogui.press('m') #mute or un_mute
                time.sleep(.15)
                
            elif(predicted_value ==21):
                final_label ="Thumb Down"
                pyautogui.press('m') #mute or un_mute
                time.sleep(.15)
                
            elif(predicted_value ==22):
                final_label ="Shaking Hand"
                pyautogui.hotkey('shift', 'v') #toggle subtitle
                time.sleep(.15)
                
            elif(predicted_value ==23):
                final_label ="Stop Sign"
                pyautogui.press('s') # to pause or stop
                time.sleep(.15)
                
            elif(predicted_value ==24):
                final_label ="Drumming Fingers"
                pyautogui.press('Space') #to play or pause
                time.sleep(.15)
                
            elif(predicted_value ==25):
                final_label ="No gesture"
                time.sleep(.15)
                
            else:
                final_label ="Doing other things"
            cv2.imshow('frame',frame)
            buffer = []
        i=i+1
        text = "activity:"+final_label
        cv2.putText(frame,text, (20,35),cv2.FONT_HERSHEY_SIMPLEX,1.155,(0, 255, 0),5)
        cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```
